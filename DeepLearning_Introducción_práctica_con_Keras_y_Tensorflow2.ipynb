```python
import tensorflow as tf
import numpy as np

```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float).reshape(-1,1)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float).reshape(-1,1)
```

```python
valor=celsius.shape[1]
print(valor)
```

```python
flatten=tf.keras.layers.Flatten(input_shape=(celsius.shape[1],))
oculta1 = tf.keras.layers.Dense(units=3)
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([flatten,oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print("Hagamos una predicción!")
resultado = modelo.predict([100.0])
print("El resultado es " + str(resultado) + " fahrenheit!")
```

```python
print("Variables internas del modelo")
#print(capa.get_weights())
print(oculta1.get_weights())
print(oculta2.get_weights())
print(salida.get_weights())
```

```python
def transformacion_johnson(X):
    transformador_johnson = PowerTransformer(method='yeo-johnson', standardize=True).fit(X)
    datos_transformados = transformador_johnson.transform(X)
    set_printoptions(precision=3)
    print(datos_transformados[:5, :])
    datos_transformados_df = pd.DataFrame(data=datos_transformados, columns=X.columns)
    return datos_transformados_df
Xpandas_T_JOHNSON1 = transformacion_johnson(X_T_JOHNSON1)
Xpandas_T_JOHNSON1.head(2)
```

```python
def transformacion_minmax_boxcox(X):
    std_dev = X.std()
    constantes = std_dev[std_dev == 0].index
    X_sin_constantes = X.drop(constantes, axis=1)
    minmax_scaler = MinMaxScaler(feature_range=(1, 2))
    Reescalar_X_R = minmax_scaler.fit_transform(X_sin_constantes)
    print("Reescalamiento Min-Max:")
    print(Reescalar_X_R[:5])
    boxcox_transformer = PowerTransformer(method='box-cox', standardize=True)
    Reescalar_X_T_BOX = boxcox_transformer.fit_transform(pd.DataFrame(Reescalar_X_R, columns=X_sin_constantes.columns))
    print("\nTransformación Box-Cox:")
    print(Reescalar_X_T_BOX[:5])
    Xpandas_T_BOX = pd.DataFrame(data=Reescalar_X_T_BOX, columns=X_sin_constantes.columns)
    return Xpandas_T_BOX
Xpandas_T_BOX1 = transformacion_minmax_boxcox(X_T_BOX1)
Xpandas_T_BOX1.head(2)
```

### FUNCIONES DE PÉRDIDA

https://keras.io/api/losses/probabilistic_losses/#binarycrossentropy-class

Binary Crossentropy (binary_crossentropy)
Descripción: Calcula la entropía cruzada para problemas de clasificación binaria.
Aplicación: Modelos de clasificación binaria (por ejemplo, detección de spam, clasificación de enfermedades como presente o ausente).

```python
model.compile(loss='binary_crossentropy')
```

Categorical Crossentropy (categorical_crossentropy)
Descripción: Calcula la entropía cruzada para problemas de clasificación multiclase con etiquetas categóricas one-hot.
Aplicación: Modelos de clasificación multiclase (por ejemplo, clasificación de imágenes en múltiples categorías).

Vectores one-hot: Es una representación en la que la clase correcta se marca con un 1 y todas las demás clases se marcan con un 0. Por ejemplo, si tienes tres clases (0, 1, 2), las etiquetas one-hot serían:
Clase 0: [1, 0, 0]
Clase 1: [0, 1, 0]
Clase 2: [0, 0, 1]


Para esto se usa la funcion "to_categoricaly": y_train_one_hot = to_categorical(y_train, num_classes=3)

```python
model.compile(loss='categorical_crossentropy')
```

Sparse Categorical Crossentropy (sparse_categorical_crossentropy)
Descripción: Similar a categorical_crossentropy, pero las etiquetas son enteros en lugar de vectores one-hot.
Aplicación: Modelos de clasificación multiclase con etiquetas enteras.

Etiquetas enteras: En lugar de usar la representación one-hot, las etiquetas se representan simplemente como enteros que indican la clase correcta. Por ejemplo, con las mismas tres clases (0, 1, 2), las etiquetas enteras serían:
Clase 0: 0
Clase 1: 1
Clase 2: 2

```python
model.compile(loss='sparse_categorical_crossentropy')
```

Mean Squared Error (mean_squared_error)
Descripción: Calcula el promedio de los cuadrados de los errores.
Aplicación: Modelos de regresión (por ejemplo, predicción de precios de casas, predicción de temperaturas).

```python
model.compile(loss='mean_squared_error')
```

Mean Absolute Error (mean_absolute_error)
Descripción: Calcula el promedio de los valores absolutos de los errores.
Aplicación: Modelos de regresión donde los errores grandes no deben ser penalizados tan fuertemente.

```python
model.compile(loss='mean_absolute_error')
```

Mean Squared Logarithmic Error (mean_squared_logarithmic_error)
Descripción: Calcula el error cuadrático medio en el espacio logarítmico.
Aplicación: Modelos de regresión donde las diferencias relativas son más importantes que las diferencias absolutas (por ejemplo, predicciones de crecimiento exponencial).

```python
model.compile(loss='mean_squared_logarithmic_error')
```

Huber Loss (huber_loss)
Descripción: Combina las ventajas de mean_squared_error y mean_absolute_error, penalizando los errores pequeños como MSE y los grandes como MAE.
Aplicación: Modelos de regresión robustos a outliers.

```python
model.compile(loss='huber_loss')
```

Kullback-Leibler Divergence (kullback_leibler_divergence o kl_divergence)
Descripción: Mide la diferencia entre dos distribuciones de probabilidad.
Aplicación: Modelos generativos como VAE (Variational Autoencoders), aplicaciones de aprendizaje no supervisado.

```python
model.compile(loss='kullback_leibler_divergence')
```

Poisson Loss (poisson)
Descripción: Calcula la pérdida basada en la distribución de Poisson.
Aplicación: Modelos donde las salidas siguen una distribución de Poisson (por ejemplo, conteos de eventos en un intervalo de tiempo fijo).

```python
model.compile(loss='poisson')
```

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#00FF00")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('black')
```

### CAPAS DE DROPOUT Y BATCHNORMALIZATION

La capa de dropout sirve para evitar el sobreajuste. Desactiva de forma aleatorea ciertas redes durante
el entrenamiento.
La capa de batch normaliza las entradas de la capa de tal manera que tengan una activación de 
salida media de cero y desviación de uno.

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras import Sequential

convolution1 = tf.keras.layers.Conv2D(filters=2*filtros, kernel_size=(3, 3), activation='relu',padding='same',input_shape=(28, 28, 1))
batch1=tf.keras.layers.BatchNormalization()
pooling1=tf.keras.layers.MaxPooling2D(pool_size=(2, 2))
convolution2 = tf.keras.layers.Conv2D(filters=4*filtros, kernel_size=(3, 3), activation='relu',padding='same')
batch2=tf.keras.layers.BatchNormalization()
pooling2=tf.keras.layers.MaxPooling2D(pool_size=(2, 2))
oculta1 = tf.keras.layers.Flatten()
oculta2= tf.keras.layers.Dense(units=64,activation='relu')
batch3=tf.keras.layers.BatchNormalization()
drop=tf.keras.layers.Dropout(0.5)
salida = tf.keras.layers.Dense(units=10,activation='softmax')
model = tf.keras.Sequential([convolution1,batch1,pooling1,convolution2,
                             batch2,pooling2,oculta1,oculta2,batch3,drop,salida])
model.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
reduce_lr = tf.keras.callbacks.LearningRateScheduler(lambda x: 1e-3 * 0.9 ** x)
model.fit(train_images, train_labels, epochs=20,callbacks=[reduce_lr])
test_loss, test_acc = model.evaluate(test_images,  test_labels)
print('Test accuracy:', test_acc)
```
